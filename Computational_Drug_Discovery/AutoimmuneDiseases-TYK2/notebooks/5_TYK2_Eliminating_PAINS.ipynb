```python
df_rd_filter = pd.read_csv('../pipeline/5_PAINS_rd_filter/all.csv')
df_rd_filter.head()
```

```python
#identify which filters remove the largest number of molecules

from collections import Counter

def counter(df):
    count_list = list(Counter(df.FILTER).items())
    count_df = pd.DataFrame(count_list, columns=['Rule', 'Count'])
    count_df.sort_values('Count', inplace=True, ascending=False)
    return count_df

counter(df_rd_filter)
```
