### Download scRNA-seq data from 10X Genomics

[Single Cell Gene Expression Dataset](https://support.10xgenomics.com/single-cell-gene-expression/datasets/3.0.0/pbmc_1k_v3)

#### 1k Peripheral Blood Mononuclear Cells (PBMCs) from a Healthy Donor (v3 chemistry)
#### Single Cell Gene Expression Dataset by Cell Ranger 3.0.0

The PBMCs were processed by 10X Genomics Chromium and sequenced on Illumina NovaSeq. 
The reads were processed using Cellranger (version 3.0.0).
PBMCs are primary cells with relatively small amounts of RNA (~1pg RNA/cell).

- 1,222 cells detected
- Sequenced on Illumina NovaSeq with approximately 54,000 reads per cell
- 28bp read1 (16bp Chromium barcode and 12bp UMI), 91bp read2 (transcript), and 8bp I7 sample barcode
