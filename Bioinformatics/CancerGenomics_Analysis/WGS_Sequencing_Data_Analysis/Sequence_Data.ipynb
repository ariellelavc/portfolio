## Convert .sra file to binary alignment/map .bam file

```python
! sam-dump --header SRR2130004/SRR2130004.sra | samtools view -bS - > BRCA_IDC_cfDNA.bam
```

## Index the .bam file

```python
! samtools index BRCA_IDC_cfDNA.bam
```

## View header information of the .bam file 

```python
! samtools view -H BRCA_IDC_cfDNA.bam
```

@HD  The header line
VN: format version
SO: Sorting order of alignments (unknown, unsorted, coordinate, queryname)

@SQ  Reference sequence dictionary
SN: reference sequence name
LN: reference sequence length

@RG  Read group
ID: read group identifier (must be unique)
PL: platform or technology used (e.g. ILLUMINA)
SM: sample ID and/or pool being sequenced

@PG  Program/tool information
ID: unique name
PN: program name
VN: program version
CL: command line

Note: The .bam file is already sorted by chromosome coordinates (`SO:coordinate`) so there's no need to sort the file using `samtools sort` prior to indexing the .bam file

```python
import pandas as pd

bam_df = pd.read_csv("aligned.paired.reads.tsv", sep="\t", header=None)
bam_df.drop(bam_df.columns[[6, 12, 13]], axis=1, inplace=True)
bam_df.columns = ['qname', 'flag', 'rname', 'pos', 'mapq', 'cigar', 'mpos', 'isize', 'seq', 'qual', 'tag']
bam_df
```

```python
bam_fields = pd.Series(bam_df.columns, name='bam fields')
bam_field_desc = pd.Series(['read/query template name', 
                            'bitwise flag describing the read alignment QC',
                            'reference sequence name (e.g chr12 or 12)',
                            'position of aligned read (leftmost coordinate)',
                            'mapping quality of the read alignment',
                            'CIGAR string: read alignment seq match to ref',
                            'mate read aligned position',
                            'template length; +/- forward/reverse strand',
                            'sequence of mapped reads on forward strand',
                            'base qualities of the sequence alignment',
                            'read group tag'
                           ],
                           name='description')
desc_df =  pd.concat([bam_fields, bam_field_desc], axis=1)
desc_df
```
