## Data preprocessing with fastp

fastp - fast all-in-one preprocessing for fastq files; performs quality control, adapter trimming, quality filtering, per-read quality pruning and many other operations with a single scan of the fastq data

```python
mkdir preprocessing_fastp
```

```python
! fastp --help
```

```python
for acc in dat:
    ! fastp -i {acc}_1.fastq.gz -I {acc}_2.fastq.gz \
            -o preprocessing_fastp/{acc}_1.fq.gz -O preprocessing_fastp/{acc}_2.fq.gz \
            -w 8 -h preprocessing_fastp/{acc}.html -j preprocessing_fastp/{acc}.json 
```

```python
ls -ls preprocessing_fastp
```

## RNA-seq reads alignment to reference genome with TopHat

TopHat uses Bowtie as the alignment engine for short reads and requires samtools as it outputs and handles alignments in binary alignment/mapping (BAM) format

```python
! tophat --version
```

```python
! bowtie2 --version
```

```python
! samtools --version
```

```python
mkdir -p ../../analysis/tophat
```

```python
for acc in dat:
    ! tophat -p 8 -G genes.gtf -o ../../analysis/tophat/{acc}_thout genome {acc}_1.fq.gz {acc}_2.fq.gz 
```
